```python
import pandas as pd
```

## Preprocessing data

```python
all_grades_form = pd.read_csv('../Resources/Datasets/General_factor_data/U.S._All_Grades_All_Formulations_Retail_Gasoline_Prices.csv', skiprows=4)
all_grades_form.head()
```

```python
all_grades_form.isnull().sum()
```

```python
all_grades_form.dtypes
```

```python
all_grades_form.columns
```

```python
#rename columns
all_grades_form = all_grades_form.rename(columns={'Month':'Period',
     'U.S. All Grades All Formulations Retail Gasoline Prices Dollars per Gallon': 'All_grades_formulation_price($/Gallon)'})

all_grades_form
```

```python
all_grades_form = all_grades_form.sort_values(by=['Period'],ascending=True)
all_grades_form
```

```python
all_grades_form = all_grades_form.reset_index()
all_grades_form
```

```python
all_grades_form = all_grades_form.drop(columns='index')
all_grades_form
```

```python
#convert to datime
all_grades_form['Period'] = pd.to_datetime(all_grades_form['Period'])
all_grades_form
```

```python
#split data
all_grades_form['Year'] = all_grades_form['Period'].dt.year
all_grades_form['Month'] = all_grades_form['Period'].dt.month
all_grades_form.head()
```

```python
#reorder columns
All_grades_form_price_df = pd.DataFrame(data = all_grades_form,
                                        columns={'Year':'Year',
                                                 'Month':'Month',
                                                'All_grades_formulation_price($/Gallon)': 'All_grades_formulation_price($/Gallon)'})

All_grades_form_price_df
```

```python
All_grades_form_price_df = All_grades_form_price_df[(All_grades_form_price_df.Year >= int(2001))]
All_grades_form_price_df
```
